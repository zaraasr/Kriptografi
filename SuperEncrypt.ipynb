```python
import numpy as np

def permute_text(plaintext, keystream):
    indices = np.argsort(keystream)
    permuted_text = ''.join([plaintext[i] for i in indices])
    return permuted_text, indices

def xor_encrypt_text_logistic_map(permuted_text, keystream):
    encrypted_text = ''.join([chr(ord(permuted_text[i]) ^ int(keystream[i % len(keystream)])) for i in range(len(permuted_text))])
    return encrypted_text

def xor_decrypt_text_logistic_map(encrypted_text, keystream):
    decrypted_text = ''.join([chr(ord(encrypted_text[i]) ^ int(keystream[i % len(keystream)])) for i in range(len(encrypted_text))])
    return decrypted_text

def inverse_permute_text(xor_decrypted_text, indices):
    original_text = ''.join([xor_decrypted_text[i] for i in np.argsort(indices)])
    return original_text, np.argsort(indices)

plaintext_text = "Hello World"
keystream = np.random.randint(256, size=len(plaintext_text))

# Enkripsi
permuted_text, indices = permute_text(plaintext_text, keystream)
encrypted_text = xor_encrypt_text_logistic_map(permuted_text, keystream)

# Dekripsi
xor_decrypted_text = xor_decrypt_text_logistic_map(encrypted_text, keystream)
decrypted_text, idx_decryption= inverse_permute_text(xor_decrypted_text, indices)

print("Nama : Zahrah Asri Nur Fauzyah")
print("NIM  : A11.2021.13769")
print("")
print("Original Text :", plaintext_text)
print("Encrypted (Permuted) Text :", permuted_text)
print("Encrypted (Permuted+Substituted) Text :", encrypted_text)
print("Decrypted (Desubstituted) Text :", xor_decrypted_text)
print("Final Decrypted Text :", decrypted_text)
```
