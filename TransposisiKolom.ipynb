```python
plainTeks = input("masukkan kata : ")
```

```python
key = input("masukkan kunci : ")
```

Membersihkan teks dari spasi ganda

```python
cleanPlainTeks = plainTeks.replace(" ", "")
print(cleanPlainTeks)
```

Deklarasi

```python
# Deklarasi urutan huruf dalam alfabet
ascii = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6, 'g': 7, 'h': 8, 'i': 9, 'j': 10, 'k': 11, 'l': 12, 'm': 13,
         'n': 14, 'o': 15, 'p': 16, 'q': 17, 'r': 18, 's': 19, 't': 20, 'u': 21, 'v': 22, 'w': 23, 'x': 24, 'y': 25, 'z': 26}
```

Mengonversi kata kunci menjadi urutan numerik sesuai dengan urutan huruf dalam deklarasi yang disediakan

```python
kolom = 0
# Inisialisasi list untuk menyimpan urutan numerik dari setiap huruf alfabet
angka_urutan = []

# Loop melalui setiap karakter dalam kata kunci
for huruf in key:
    # Periksa apakah karakter saat ini adalah huruf alfabet
    if huruf.isalpha():
        # Dapatkan urutan numerik dari huruf tersebut dan tambahkan ke dalam list angka_urutan
        angka_urutan.append(ascii[huruf.lower()])
        kolom = kolom+1
        cleanKey = angka_urutan
# Cetak urutan numerik dari kata kunci
print(f"Urutan kunci dari kata '{key}' adalah: {angka_urutan}")
```

Menghitung jumlah baris dan jumlah kolom yang diperlukan untuk membuat matriks berdasarkan panjang teks yang sudah dibersihkan, Pembulatan ini memastikan bahwa matriks memiliki ukuran yang tepat untuk menampung semua elemen teks yang akan dimasukkan ke dalamnya, sehingga tidak ada elemen teks yang terpotong atau tidak masuk ke dalam matriks.

```python
# Membuat sebuah matriks kosong untuk menyimpan teks yang telah diolah dalam bentuk matriks.
matrix = []
#Menghitung panjang dari teks yang telah dibersihkan dan disimpan dalam variabel cleanPlainTeks.
panjang = len(cleanPlainTeks)
#Mencari tahu berapa kali kolom dapat digunakan untuk memuat seluruh teks, dengan membagi panjang teks dengan jumlah kolom yang mungkin.
temp = panjang/kolom
#Membulatkan hasil pembagian sebelumnya ke bilangan bulat terdekat.
bulat = round(temp)
#Memeriksa apakah sisa dari pembagian sebelumnya negatif. Ini terjadi jika jumlah kolom yang diinginkan lebih besar dari panjang teks.
if temp - bulat == 0:
    baris = bulat #Jika sisa dari pembagian adalah negatif, maka jumlah baris yang dibutuhkan tetap sama dengan nilai yang telah dibulatkan sebelumnya.
else:
    if temp - bulat < 0:
        baris = bulat
    else:
        baris = bulat + 1
print(baris)
```

Mengonversi teks yang telah dibersihkan (cleanPlainTeks) menjadi sebuah matriks dengan jumlah baris dan kolom yang telah ditentukan sebelumnya.

```python
for i in range(baris):
    # Membuat list kosong untuk setiap baris dalam matriks
    row = []
    for j in range(kolom):
        if i * kolom + j < len(cleanPlainTeks):
            # Memeriksa apakah indeks saat ini berada dalam rentang panjang teks yang dibersihkan
            elemen = cleanPlainTeks[i * kolom + j]
        else:
            # Jika indeks melebihi panjang teks, mengisi elemen dengan spasi
            elemen = " "
        # Menambahkan elemen ke dalam baris
        row.append(elemen)
    matrix.append(row)
```

```python
print("Matriks yang Anda buat:")
for row in matrix:
    print(row)
```

Mentranspose matriks yang telah dibuat sebelumnya. Proses transpose ini mengubah baris menjadi kolom dan sebaliknya

```python
transposed_matrix = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
for row in transposed_matrix:
    print(row)
```

Mengurutkan kunci

```python
# Membuat list kosong dengan panjang yang sama dengan cleanKey untuk menyimpan urutan indeks
order = [None] * len(cleanKey)
i = 0
while len(cleanKey) > 0:
    # Membuat list nilai yang bukan None dari cleanKey
    non_none_values = list(filter(lambda x: x is not None, cleanKey))
    if non_none_values:
        ## Mencari nilai terkecil dari list nilai yang bukan None
        nilai_terkecil = min(non_none_values)
        marker = cleanKey.index(nilai_terkecil)
        ## Menyimpan nomor urut ke indeks sesuai dengan nilai terkecil
        order[marker] = i
        cleanKey[marker] = None  # Hapus nilai terbesar dari array
        i = i + 1
    else:
        break
print(order)
```

mengubah urutan kolom dalam matriks transposed_matrix berdasarkan urutan yang telah ditentukan sebelumnya dalam list order.

```python
new_matrix = []
while len(order) > 0:
    non_none_values = list(filter(lambda x: x is not None, order))
    if non_none_values:
        # Mencari nilai terkecil dari non_none_values
        nilai_terkecil = min(non_none_values)
        marker = order.index(nilai_terkecil)
        # Menambahkan baris dengan indeks yang sesuai ke new_matrix
        new_matrix.append(transposed_matrix[marker])
        order[marker] = None
    else:
        break

# Menampilkan matriks yang sudah diubah urutan kolomnya
for row in new_matrix:
    print(row)
```

```python
# Inisialisasi string kosong
result_string = ""

# Iterasi melalui matriks dan menggabungkan nilai-nilainya ke dalam string
for row in new_matrix:
    for value in row:
        result_string += str(value)

# Menghapus spasi ekstra di akhir string
result_string = result_string.strip()

# Menampilkan hasil
print(result_string)
```
