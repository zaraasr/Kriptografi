#TRANSPOSISI RAIL FENCE

Input Plaintext

```python
plainText = input("Plaintext : ")
```

Menentukkan Jumlah Baris (kunci)

```python
railFenceKey = int(input("Kunci: "))
```

Membuat Matriks Kosong Sesuai Dengan Jumlah Baris (kunci)

```python
#Setiap elemen dalam matriks diisi dengan spasi kosong, kemudian diisi dengan huruf-huruf dari plaintext sesuai dengan pola zig-zag yang dibuat.
rails = [[' ' for _ in range(len(plainText))] for _ in range(railFenceKey)]
```

Menentukan Arah Pola Zig-Zag Dari Huruf-Huruf Dalam Plaintext

```python
# Insialisasi untuk membuat pola zigzag
i, j = 0, 0
direction = 1  #jika 1 maka pola turun, jika -1 maka pola naik

for char in plainText:
  rails[i][j] = char
  if i == 0:
    direction = 1
  elif i == railFenceKey - 1:
    direction = -1
  i += direction

  j += 1
```

Membentuk Teks Terenkripsi Berdasarkan Pola Zig-Zag

```python
#menggabungkan semua karakter dalam satu baris matriks menjadi satu string
enkripsi_teks = '\n'.join([''.join(row) for row in rails])
print("* RAIL FENCE: ")
print(enkripsi_teks)
```

Mempersiapkan struktur data(Kotak) yang diperlukan untuk menyimpan karakter-karakter menggunakan algoritma Rail Fence.

```python
#membuat array rails yang terdiri dari railFenceKey sublist kosong. Setiap sublist akan digunakan untuk menyimpan karakter-karakter yang akan disusun sesuai dengan pola zigzag.
rails = [[] for _ in range(railFenceKey)]
#menunjukkan indeks sublist yang sedang diisi saat ini
index = 0
#menentukan arah perpindahan. jika 1 maka turun, jika -1 maka naik
direction = 1
```

Mengisi karakter-karakter dari plaintext ke dalam sublist yang sesuai dalam struktur data rails, sesuai dengan pola zigzag yang digunakan dalam algoritma Rail Fence.

```python
#iterasi dilakukan pada tiap karakter dalam plaintext, setiap karakter ditambahkan ke sublist yang sesuai dengan index saat ini.
for char in plainText:
  rails[index].append(char)
  #jika index = 0, arah nya akan berubah kebawah
  if index == 0:
    direction = 1
  #jika sudah mencapai index terbesar (railFenceKey-1), arah nya akan berubah keatas
  elif index == railFenceKey - 1:
    direction = -1
  #menambahkan index dengan direction, untuk menentukan apakah ke atas atau ke bawah
  index += direction
```

**Operasi "join(rail) for rail in rails"** untuk menggabungkan karakter-karakter dalam setiap rail menjadi satu string.

**Operasi "join(...)"** untuk menggabungkan semua string hasil tersebut menjadi satu string tunggal.







```python
#melakukan join pada setiap rail dalam array rails
enkripsi_teks = "".join(["".join(rail) for rail in rails])
print("CIPHERTEXT : " + enkripsi_teks)
```

```python
#Mengisi placeholder yang sudah dibuat dengan teks enkripsi
index_char = 0
for rail in rails:
  for i in range(len(rail)):
    rail[i] = enkripsi_teks[index_char]
    index_char += 1
```

Proses Dekripsi

```python
#mendekrasikan ulang, mengambil karakter-karakter dari "kotak-kotak" sesuai dengan pola yang telah ditentukan.
index = 0
direction = 1
dekripsi_teks = []

for _ in range(len(enkripsi_teks)):
  #pop(0) digunakan untuk menghapus karakter pertama yang sesuai index pada rails, dan dibangun ulang pada array deskripsi_teks
  dekripsi_teks.append(rails[index].pop(0))
  #jika index = 0, arah nya akan berubah kebawah
  if index == 0:
    direction = 1
  #jika sudah mencapai index terbesar (railFenceKey-1), arah nya akan berubah keatas
  elif index == railFenceKey - 1:
    direction = -1
  index += direction
```

Menggabungkan karakter-karakter dalam list dekripsi_teks menjadi sebuah string, yang merupakan pesan yang telah didekripsi kembali.

```python
railFenceDecryption = "".join(dekripsi_teks)
print(" DECRYPTED : " + railFenceDecryption)
```
